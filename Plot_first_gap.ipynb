```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
q5 = np.load("q5_extr.npy")
q7 = np.load("q7_extr.npy")
q9 = np.load("q9_extr.npy")
J = np.linspace(0.,1,30)
```

```python
plt.plot(J, q5, label = "L = 4")
plt.plot(J, q7, label = "L = 6")
plt.plot(J, q9, label = "L = 8")
plt.plot(J, 2*(1-J), label = "Exact solution")
plt.xlabel("J")
plt.ylabel(r"$\Delta$")
plt.xlim(0, 1)
plt.ylim(0, 2)
plt.legend()
plt.savefig("First_gaps.pdf")
plt.savefig("First_gaps.svg")
```
